#### graph function

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

##### helper function below

```python
def inclusive_prediction_scores(model_name):
  result_path = "../../results/"+model_name+".csv"
  bart_large_mnli=pd.read_csv(result_path).sample(frac=1)
  fig, ax = plt.subplots(figsize=(8, 4))
  ax.set_title(model_name+' : inclusive class prediction scores')
  ax.set_xlabel('Exclusive <---> Inclusive')
  ax.yaxis.set_major_locator(mpl.ticker.NullLocator())
  ax.plot(list(bart_large_mnli['inclusive']), np.arange(0,len(bart_large_mnli['inclusive'])), 'o')
```

```python
def exclusive_prediction_scores(model_name):
  result_path = "../../results/"+model_name+".csv"
  bart_large_mnli=pd.read_csv(result_path).sample(frac=1)
  fig, ax = plt.subplots(figsize=(8, 4))
  ax.set_title(model_name+' : exclusive class prediction scores')
  ax.set_xlabel('Inclusive <---> Exclusive')
  ax.yaxis.set_major_locator(mpl.ticker.NullLocator())
  ax.plot(list(bart_large_mnli['exclusive']), np.arange(0,len(bart_large_mnli['exclusive'])), 'o', color='red')

  exclusive = list(bart_large_mnli['exclusive'])
  exclusive.sort()
  print(exclusive)
```

```python
def both_prediction_scores(model_name):
  result_path = "../../results/"+model_name+".csv"
  bart_large_mnli=pd.read_csv(result_path).sample(frac=1)
  fig, ax = plt.subplots(figsize=(8, 4))
  ax.set_title(model_name+' : prediction scores')
  ax.set_xlabel('Exclusive <---> Inclusive')
  ax.yaxis.set_major_locator(mpl.ticker.NullLocator())
  ax.plot(list(bart_large_mnli['inclusive']), np.arange(0,len(bart_large_mnli['inclusive'])), 'o')
```

### Graphs Below

```python
inclusive_prediction_scores("bart-large-mnli")
exclusive_prediction_scores("bart-large-mnli")
inclusive_prediction_scores("french_xlm_xnli")
exclusive_prediction_scores("french_xlm_xnli")
```
